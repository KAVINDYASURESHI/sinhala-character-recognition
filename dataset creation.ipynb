```python
import os
import numpy as np
```

```python
data_path='dataset'
```

```python
folders=os.listdir(data_path)
print(folders)
labels=[i for i in range(len(folders))]
print(labels)
```

```python
category_dict={'a':0,'ae':1,'e':2,'u':3}
print(category_dict)
```

```python
import cv2
data=[]
target=[]

for folders in folders:
    img_folder_path=os.path.join(data_path,folders)
    #print(img_folder_path)
    images=os.listdir(img_folder_path)
    for images in images:
        images_path=os.path.join(img_folder_path,images)
        img=cv2.imread(images_path,0)
        img=cv2.resize(img,(8,8))
        data.append(img)
        target.append(category_dict[folders])

    
```

```python
print(len(data),len(target))
```

```python
print(data[10])
print(target[10])
```

```python
data=np.array(data)
data=data.reshape(data.shape[0],data.shape[1]*data.shape[2])

target=np.array(target)

print(data.shape)
print(target.shape)
```

```python
np.save('data',data)
np.save('target',target)
```
